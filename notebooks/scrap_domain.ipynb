```python
### Now starting build content structure
n = 50 #Page limit
with open('result.csv','a') as f:
    f.write('id,price,street,suburb,state,postcode,lat,lng,beds,baths,parking,landSize,landUnit')
```

```python
for i in range(1,n+1):
    url = 'http://www.domain.com.au/rent/?region=northern-victoria-vic,south-western-victoria-vic,gippsland-greater-region-vic&excludedeposittaken=1&page='+str(i)
    r = requests.get(url, cookies = cookies, headers = headers)
    print('Now searching page: ',i)
    soup = BeautifulSoup(r.content, 'html.parser')
    a_tags = soup.find_all('script')
    content = a_tags[34].string
    data = json.loads(content)
    records = data['props']['pageProps']['layoutProps']['digitalData']['page']['pageInfo']['search']['resultsRecords']
    for record in records.split(','):
        content = record + ','
        price = data['props']['pageProps']['componentProps']['listingsMap'][record]['listingModel']['price']
        address = data['props']['pageProps']['componentProps']['listingsMap'][record]['listingModel']['address']
        feature = data['props']['pageProps']['componentProps']['listingsMap'][record]['listingModel']['features']
        f_k = feature.keys()
        if 'parking' not in f_k:
            feature['parking'] = 0
        if 'beds' not in f_k:
            feature['beds'] = 0
        if 'baths' not in f_k:
            feature['baths'] = 0
        if 'landSize' not in f_k:
            feature['landSize'] = 0
        #print(feature)
        content+=price+','
        content+=address['street']+','
        content+=address['suburb']+','
        content+=address['state']+','
        content+=address['postcode']+','
        content+=str(address['lat'])+','
        content+=str(address['lng'])+','
        content+=str(feature['beds'])+','
        content+=str(feature['baths'])+','
        content+=str(feature['parking'])+','
        content+=str(feature['landSize'])+','
        content+=feature['landUnit']
        content+='\n'
        with open('result.csv','a') as f:
            f.write(content)
```

```python
import pandas as pd
import folium # This is a library to plot map.
```

```python
data = pd.read_csv('result.csv')
data
```

```python
# San Francisco latitude and longitude values
latitude = -37.477
longitude = 144.945

# Create map and display it
san_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# Display the map of San Francisco
san_map
```

```python
# get the first 200 records
limit = 200
cdata = data.iloc[0:limit, :]

# Instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# Loop through the 200 crimes and add each to the incidents feature group
for lat, lng, in zip(cdata.lat, cdata.lng):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )

# Add incidents to map
san_map = folium.Map(location=[latitude, longitude], zoom_start=12)
san_map.add_child(incidents)
```
