```python
# Get all pages of the sitemap of the website that contains the suburbs of Victoria
# See: https://www.oldlistings.com.au/site-map?state=VIC


def getlinks(base_url,seed,class_str,ua_l):
    
    gateway = get_gateway(base_url);
    
    link_l = [seed];
    visited = [];
    notvisited = [seed];

    while notvisited !=[]:
        
        site_str = notvisited[0]

        page = getpage_aws_ip(gateway,base_url,site_str,ua_l)
    
        visited.append(site_str);
        notvisited.pop(0);
        
        soup = BeautifulSoup(page.content, 'html.parser');
        prop_div= soup.find_all("li", {"class":[class_str]});
        
        
        for tag in prop_div:
            link_str =tag.find('a').get('href');
            if link_str not in link_l:
                link_l.append(link_str);
                notvisited.append(link_str);
                
    gateway.shutdown() 
                
    return link_l
```

```python
base_url = 'https://www.oldlistings.com.au';   

ua_l = get_useragent();

sitemap_seed = '/site-map?state=VIC';
sitemap_class = 'pager-next'
sitemaplinks_l = getlinks(base_url,sitemap_seed,sitemap_class,ua_l)
```
